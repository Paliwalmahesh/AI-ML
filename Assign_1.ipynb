```python
import math
import pandas as pd 
import numpy as np
filepath = r"D:\Study\College\Engneering\3rd Year\#CS_Lab\infy data.csv"
df = pd.read_csv(filepath)

#1 Calculate mean, variance, standard deviation, and standard error of Close Price 

#Mean
data=list(df.loc[:,'Close Price'])
mean=0
for n in range(len(data)):
    mean=mean+data[n]
mean=mean//len(data)
print("Mean of Close Price=",mean)

#Variance()
variance=0
for n in range(len(data)):
    variance=variance+((data[n]-mean)*(data[n]-mean))
variance=variance//(len(data)-1)
print("Variance of Close Price=",variance)

#standard deviation
std_dev=math.sqrt(variance//(len(data))) 
print("Standard Deviation of Close Price=",int(std_dev))

#standard error
std_error=std_dev//math.sqrt(len(data))
print("Standard Error of Close Price=",std_error)




```

```python
#2.	Calculate Covariance and Correlation between Open Price and Close Price.

#Covariance
covariance=0
data_1=list(df.loc[:,'Open Price'])

mean_1=0
for n in range(len(data_1)):
    mean_1=mean_1+data_1[n]
mean_1=mean_1/len(data_1)

for n in range(len(data)):
    covariance=covariance+((data_1[n]-mean_1)*(data[n]-mean))
covariance=covariance//len(data)
print("Covariance between Open Price and Close Price=",covariance)

#Correlation
covariance=0
a=0
a1=0
for n in range(len(data)):
    covariance=covariance+((data_1[n]-mean_1)*(data[n]-mean))
    a=a+((data_1[n]-mean_1)**2)
    a1=a1+((data[n]-mean)**2)
correlation=covariance/(math.sqrt(a*a1))
print("Correlation between Open Price and Close Price=",correlation)

```
